```python
import math
from poisson_process import CompoundPoisson, CramerLundberg, MixtureDistribution, NonHomogeneousPoissonProcess
from poisson_process import PoissonProcess1, PoissonProcess2, PoissonProcess3
from poisson_process import torch, plt , quad,norm
from tqdm import tqdm
```

```python
# Parámetros
n_sim = 10000
t = 2.0
lam = 3.0

lam_hat = torch.zeros(n_sim)
running_sum = 0

poi_pro = PoissonProcess3(lam)

for i in tqdm(range(n_sim)):
    
    # Simular N(t)
    N_t = poi_pro.simulate(t)[1][-1]
    
    running_sum += N_t
    lam_hat[i] = running_sum / (i + 1)

# Valor teórico
true_value = lam 

# Gráfica
plt.figure()
plt.plot(lam_hat.numpy()/t, label="Estimador Monte Carlo",color='red')
plt.axhline(true_value, linestyle="--", label="Valor teórico λ")
plt.xlabel("Número de simulaciones")
plt.ylabel("Estimación de λ")
plt.title("Convergencia Monte Carlo")
plt.legend()
plt.show()
```

```python
# Parámetros
n_sim = 100000
t = 12.0

def lam1(x):
    return 2 + math.sin(x)

nh_poi_pro = NonHomogeneousPoissonProcess(lam1)

lam_hat = torch.zeros(n_sim)
running_sum = 0

for i in tqdm(range(n_sim)):
    
    # Simular N(t)
    N_t = nh_poi_pro.simulate(t,10)[1][-1]
    
    running_sum += N_t
    lam_hat[i] = running_sum / (i + 1)

# Valor teórico
true_value = quad(lam1,0,t)[0]

# Gráfica
plt.figure()
plt.plot(lam_hat.numpy(), label="Estimador Monte Carlo",color='red')
plt.axhline(true_value, linestyle="--", label="Valor teórico E[N(t)]")
plt.xlabel("Número de simulaciones")
plt.ylabel("Estimación de E[N(t)]")
plt.title("Convergencia Monte Carlo")
plt.legend()
plt.show()
```

```python
lam = 2
n_sim = 100000
t = 10
beta = 2.5

# saltos exponenciales
jump_dist = torch.distributions.Exponential(beta)

cp = CompoundPoisson(lam, jump_dist)

lam_hat = torch.zeros(n_sim)
running_sum = 0

for i in tqdm(range(n_sim)):
    
    # Simular X(t)
    X_t = cp.simulate(t)[1][-1]
    
    running_sum += X_t
    lam_hat[i] = running_sum / (i + 1)

# Valor teórico
true_value = 2

lam_hat = beta*lam_hat/t
# Gráfica
plt.figure()
plt.plot(lam_hat.numpy(), label="Estimador Monte Carlo",color='red')
plt.axhline(true_value, linestyle="--", label="Valor teórico λ")
plt.xlabel("Número de simulaciones")
plt.ylabel("Estimación de λ")
plt.title("Convergencia Monte Carlo")
plt.legend()
plt.show()
```

```python
lam = 5
u = 200
beta = 1/10
i = 0.05
c = (1+i)*lam*(1/beta)
t = 12*5
n_sim = 100000

# saltos exponenciales
claim_dist = torch.distributions.Exponential(beta)

cl_p = CramerLundberg(u, c, lam, claim_dist)

lam_hat = torch.zeros(n_sim)
running_sum = 0

for i in tqdm(range(n_sim)):
    
    # Simular U(t)
    U_t = cl_p.simulate(t)[1][-1]
    
    running_sum += U_t
    lam_hat[i] = running_sum / (i + 1)

# Valor teórico
true_value = 5

lam_hat = beta*(u +c*t -lam_hat)/t
# Gráfica
plt.figure()
plt.plot(lam_hat.numpy(), label="Estimador Monte Carlo",color='red')
plt.axhline(true_value, linestyle="--", label="Valor teórico λ")
plt.xlabel("Número de simulaciones")
plt.ylabel("Estimación de λ")
plt.title("Convergencia Monte Carlo")
plt.legend()
plt.show()
```

```python
def ruin_convergence(model, T, n_sim, alpha=0.05):
    """
    Grafica la convergencia del estimador Monte Carlo
    de la probabilidad de ruina.
    
    Parámetros
    ----------
    model : CramerLundberg
        Modelo ya inicializado.
        
    T : float
        Horizonte temporal.
        
    n_sim : int
        Número total de simulaciones.
        
    alpha : float
        Nivel para bandas tipo CLT.
    """
    
    estimates = torch.zeros(n_sim)
    running_sum = 0
    
    for i in tqdm(range(n_sim)):
        
        ruined = model.simulate(T)[2]
        
        running_sum += ruined
        p_hat = running_sum / (i + 1)
        
        estimates[i] = p_hat

    # Intervalo tipo CLT dinámico
    z = norm.ppf(1 - alpha/2)
    
    n_vals = torch.arange(1, n_sim + 1)
    se = torch.sqrt(estimates * (1 - estimates) / n_vals)
    
    lower = estimates - z * se
    upper = estimates + z * se

    # Gráfica
    plt.figure()
    plt.plot(estimates.numpy(), label="Estimador Monte Carlo")
    plt.plot(lower.numpy(), linestyle="--", label="IC inferior")
    plt.plot(upper.numpy(), linestyle="--", label="IC superior")
    plt.fill_between(
        n_vals,
        lower,
        upper,
        alpha=0.3,
        label=f"IC {(1-alpha)*100:.0f}%"
    )
    
    plt.xlabel("Número de simulaciones")
    plt.ylabel("Probabilidad de ruina estimada")
    plt.title("Convergencia Monte Carlo - Probabilidad de Ruina")
    plt.legend()
    plt.show()
```

```python
lam = 5
u = 200
beta = 1/10
alpha = 2
i = 0.1
c = (1+i)*lam*(alpha/beta)
t = 12*5

# saltos exponenciales
claim_dist = torch.distributions.Gamma(alpha,beta)

cl_p = CramerLundberg(u, c, lam, claim_dist)

ruin_convergence(cl_p, T=t, n_sim=100000)
```

```python
P = cl_p.ruin_probability(t,10000)
print(f"Probabilidad de ruina: {P[0]},\nIntervalo del 95% de confianza: {P[1:]}")
```

```python
lam = 5
u = 200
beta = 1/10
alpha = 2
i = -0.1
c = (1+i)*lam*(alpha/beta)
t = 12*5

# saltos exponenciales
claim_dist = torch.distributions.Gamma(alpha,beta)

cl_p = CramerLundberg(u, c, lam, claim_dist)

ruin_convergence(cl_p, T=t, n_sim=100000)
```

```python

alphas = [2.0,8.0]
betas = [2.0,0.5]
gamma1 = torch.distributions.Gamma(concentration=alphas[0], rate=betas[0])
gamma2 = torch.distributions.Gamma(concentration=alphas[1], rate=betas[1])

densities = [gamma1,gamma2]
weights = torch.Tensor([0.7,0.3])
lam = 10
u = 200
beta = 1/10
alpha = 2
i = 0.08
c = (1+i)*lam*((weights[0]*alphas[0]/betas[0])+weights[1]*alphas[1]/betas[1])
t = 12*5

# saltos exponenciales
claim_dist = MixtureDistribution(weights,densities)

cl_p = CramerLundberg(u, c, lam, claim_dist)
cl_p.plot(t,linestyle="-")
```

```python
def ruin_vs_u(model, u_values, T, n_sim):

    probs = []

    for u in tqdm(u_values, desc="Variando u"):
        
        model.u = float(u)
        ruins = 0
        
        for _ in range(n_sim):
            ruins += model.simulate(T)[2]
        
        probs.append(ruins / n_sim)

    plt.figure()
    plt.plot(u_values, probs)
    plt.xlabel("Capital inicial u")
    plt.ylabel("Probabilidad de ruina")
    plt.title("Ruina vs Capital inicial")
    plt.grid(alpha=0.3)
    plt.show()
```

```python
lam = 5
u = 200
beta = 1/10
alpha = 2
i = 0.1
c = (1+i)*lam*(alpha/beta)
t = 12*5

# saltos exponenciales
claim_dist = torch.distributions.Gamma(alpha,beta)

cl_p = CramerLundberg(u, c, lam, claim_dist)

u_vals = torch.linspace(0,1000,10)
ruin_vs_u(cl_p,u_vals,t,10000)
```

```python
def ruin_vs_c(model, c_values, T, n_sim):

    probs = []

    for c in tqdm(c_values, desc="Variando c"):
        
        model.c = float(c)
        ruins = 0
        
        for _ in range(n_sim):
            ruins += model.simulate(T)[2]
        
        probs.append(ruins / n_sim)

    plt.figure()
    plt.plot(c_values, probs)
    plt.xlabel("Prima c")
    plt.ylabel("Probabilidad de ruina")
    plt.title("Ruina vs Prima")
    plt.grid(alpha=0.3)
    plt.show()
```

```python
c_vals = torch.linspace(0,1000,100)
ruin_vs_c(cl_p,c_vals,t,10000)
```

```python
def ruin_vs_T(model, T_values, n_sim):

    probs = []

    for T in tqdm(T_values, desc="Variando T"):
        
        ruins = 0
        
        for _ in range(n_sim):
            ruins += model.simulate(T.item())[2]
        
        probs.append(ruins / n_sim)

    plt.figure()
    plt.plot(T_values, probs)
    plt.xlabel("Horizonte T")
    plt.ylabel("Probabilidad de ruina")
    plt.title("Ruina vs Horizonte")
    plt.grid(alpha=0.3)
    plt.show()
```

```python
lam = 5
u = 200
beta = 1/10
alpha = 2
i = 0.1
c = (1+i)*lam*(alpha/beta)
t = 12*5

# saltos exponenciales
claim_dist = torch.distributions.Gamma(alpha,beta)

cl_p = CramerLundberg(u, c, lam, claim_dist)


t_vals = torch.linspace(12,600,100)
ruin_vs_T(cl_p,t_vals,10000)
```
