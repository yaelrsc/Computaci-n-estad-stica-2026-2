```python

class EmpiricalCDF:

    def __init__(self, data):

        self.data = data 
        
    def compute(self, x, alpha = 0.05, print_ = True):

        n = len(data)
        Fnx = len(data[data <= x])/n
        ep = math.sqrt(math.log(2/alpha)/n)
        Conf_band = (Fnx-ep, Fnx+ep)

        if print_:
            print("Empirical CDF: {:.4f}, Confidence Band: ({:.4f}, {:.4f})".format(Fnx, Conf_band[0], Conf_band[1]))

        return Fnx, Conf_band

    def plot (self, x_min, x_max, alpha = 0.05, n_points = 1000): 
        
        delta = (x_max - x_min)/n_points

        points = torch.arange(x_min, x_max, delta)

        Fnx = torch.empty((len(points),))
        low = torch.empty((len(points),))
        up = torch.empty((len(points),))
        

        for i in range(n_points):

            Fnx[i], conf_band = self.compute(points[i],alpha=alpha,print_=False)
            low[i], up[i] = conf_band

        return Fnx, low, up
        
        
        
```

```python
import matplotlib.pyplot as plt

class EmpiricalCDF:
    def __init__(self, data):
 
        self.data = torch.sort(data).values
        self.n = len(data)
        
    def compute(self, x, alpha=0.05, print_=True):
   
        count = torch.searchsorted(self.data, x, right=True)
        Fnx = count.float() / self.n
        
        ep = math.sqrt(math.log(2/alpha) / (2 * self.n))
        low = torch.clamp(Fnx - ep, 0., 1.)
        up = torch.clamp(Fnx + ep, 0., 1.)
        
        if print_:
            print(f"ECDF: {Fnx:.4f}, Band: ({low:.4f}, {up:.4f})")
        return Fnx, (low, up)

    def plot(self, x_min, x_max, alpha=0.05, n_points=1000):
     
        points = torch.linspace(x_min, x_max, n_points)
        Fnx, (low, up) = self.compute(points, alpha=alpha, print_=False)
        
        plt.figure(figsize=(10, 6))
        plt.fill_between(points, low, up, color='red', alpha=0.2, label='Banda de Confianza')
        plt.step(points, Fnx, where='post', label='ECDF', color='blue')
        
        plt.title(f"Empirical CDF con Banda de Confianza (α={alpha})")
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.show()
```

```python
norm = torch.distributions.Normal(0,1)
n_sample = 10000
data = norm.sample((n_sample,))
norm_ecdf = EmpiricalCDF(data)
norm_ecdf.compute(0.0)
```

```python
norm_ecdf.plot(-3.,3.,n_points=10000)
```

```python
norm = torch.distributions.Uniform(0,1)
n_sample = 1000000
data = norm.sample((n_sample,))
norm_ecdf = EmpiricalCDF(data)
norm_ecdf.compute(0.5)
norm_ecdf.plot(0.1,0.2,n_points=1000)
```
