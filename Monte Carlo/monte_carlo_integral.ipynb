```python
import numpy
import math
```

```python
import torch
```

```python
Z = torch.distributions.Normal(0,1)

def monte_carlo_integral(func,a,b,n, alpha=0.05):
    unif = torch.distributions.Uniform(a,b)
    sample_ = func(unif.sample((n,)))
    integral = (b-a)*sample_.mean()
    integral_se = (b-a)*sample_.std()
    alpha = torch.Tensor([1-alpha/2])
    z_alpha = Z.icdf(alpha)
    low = integral - z_alpha*integral_se/(math.sqrt(n))
    up = integral + z_alpha*integral_se/(math.sqrt(n))
    conf_int = (low, up)
    return integral, conf_int
    
    

```

```python
def f(x):
    return x
a = 0
b = 1
monte_carlo_integral(f,a,b,10000000)
```

```python
def f(x):
    return torch.exp(x)
monte_carlo_integral(f,a,b,10000000)
```
