## **Regresión t-student**

```python
class BayesianStudentTRegression(BayesianLinearRegression):

    def __init__(self, nu_prior=None, sigma_prior=None, **kwargs):

        super().__init__(**kwargs)

        self.nu_prior = nu_prior if nu_prior is not None else dist.HalfNormal(1.0)
        self.sigma_prior = sigma_prior if sigma_prior is not None else dist.HalfNormal(1.0)

    def model(self, X, y=None):

        n, p = X.shape

        beta = numpyro.sample(
            "beta",
            self.beta_prior.expand([p]).to_event(1)
        )

        nu = numpyro.sample(
            "nu",
            self.nu_prior
        ) + 2

        sigma = numpyro.sample(
            "sigma",
            self.sigma_prior
        )

        mu = jnp.dot(X, beta)

        with numpyro.plate("data", n):

            numpyro.sample(
                "y",
                dist.StudentT(df=nu, loc=mu, scale=sigma),
                obs=y
            )

    def neg_log_likelihood(self, X, y, params):

        beta = params["beta"]
        sigma = params["sigma"]
        nu = params["nu"]

        mu = jnp.dot(X, beta)

        return -jnp.sum(dist.StudentT(df=nu, loc=mu, scale=sigma).log_prob(y))


    def sample_likelihood(self, mu, params):

        self.rng_key, subkay = random.split(self.rng_key)

        sigma = params["sigma"]
        nu = params["nu"]

        return dist.StudentT(df=nu, loc=mu, scale=sigma).sample(subkay)


```

```python
class BayesianMixtureShrinkageLinearRegression(BayesianLinearRegression):

    def __init__(self, tau_prior=None, b_prior=None, alpha_prior=None, **kwargs):

        super().__init__(**kwargs)

        self.tau_prior = tau_prior if tau_prior is not None else dist.HalfNormal(1.0)
        self.b_prior = b_prior if b_prior is not None else dist.HalfNormal(1.0)
        self.alpha_prior = alpha_prior if alpha_prior is not None else dist.Beta(1.0, 1.0)

        # IMPORTANTE: no usar beta_prior aquí
        self.beta_prior = None


    # =========================================================
    # MODEL
    # =========================================================

    def model(self, X, y=None):

        n, p = X.shape

        # hiperparámetros
        b = numpyro.sample("b", self.b_prior)
        tau = numpyro.sample("tau", self.sigma_prior)
        alpha = numpyro.sample("alpha", self.alpha_prior)

        beta_prior = dist.Mixture(dist.Categorical(probs=jnp.array([alpha, 1-alpha])), [dist.Normal(0, tau), dist.Laplace(0, b)])

        sigma = numpyro.sample("sigma", self.sigma_prior)

        beta = numpyro.sample(
            "beta",
            beta_prior.expand([p]).to_event(1)
        )

        mu = jnp.dot(X, beta)

        with numpyro.plate("data", n):

            numpyro.sample("y", dist.Normal(mu, sigma), obs=y)




```

## **Regresión con Mezcla de Gaussianas**

```python
class BayesianGaussianMixtureRegressionMHMC(BayesianLinearRegression):

    def __init__(self, K=2, pi_prior=None, **kwargs):
        super().__init__(**kwargs)

        self.K = K
        self.pi_prior = pi_prior if pi_prior is not None else dist.Dirichlet(jnp.ones(K))

    # =========================================================
    # MODEL (con variable latente z)
    # =========================================================

    def model(self, X, y=None):

        n, p = X.shape
        K = self.K

        pi = numpyro.sample("pi", dist.Dirichlet(jnp.ones(K)))

        beta = numpyro.sample(
            "beta",
            self.beta_prior.expand([K, p]).to_event(2)
        )

        sigma = numpyro.sample(
            "sigma",
            self.sigma_prior.expand([K]).to_event(1)
        )

        mu = jnp.dot(X, beta.T)   # (n, K)

        with numpyro.plate("data", n):

            mixture_dist = dist.MixtureSameFamily(
                dist.Categorical(pi),
                dist.Normal(mu, sigma)
            )

            numpyro.sample("y", mixture_dist, obs=y)

    # =========================================================
    # FIT (usa MixedHMC)
    # =========================================================

    # =========================================================
    # NLL (marginalizada para estabilidad)
    # =========================================================

    def neg_log_likelihood(self, X, y, params):

        beta = params["beta"]      # (K, p)
        sigma = params["sigma"]    # (K,)
        pi = params["pi"]          # (K,)

        mu = jnp.dot(X, beta.T)    # (n, K)

        log_probs = dist.Normal(mu, sigma).log_prob(y[:, None])

        log_mix = jax.scipy.special.logsumexp(
            jnp.log(pi) + log_probs,
            axis=1
        )

        return -jnp.sum(log_mix)

    # =========================================================
    # POSTERIOR PREDICTIVE
    # =========================================================

    def sample_likelihood(self, mu, params):
        """
        Aquí usamos z explícito (si está en samples)
        """

        self.rng_key, key = random.split(self.rng_key)

        samples = params
        beta = samples["beta"]
        sigma = samples["sigma"]
        pi = samples["pi"]

        S = beta.shape[0]
        n = mu.shape[0]

        # sample z
        z = dist.Categorical(pi).sample(key, sample_shape=(n,))  # (n, S)

        # seleccionar medias
        mu_selected = jnp.take_along_axis(mu, z[:, :, None], axis=2).squeeze(-1)
        sigma_selected = jnp.take_along_axis(sigma.T, z, axis=0).T

        return dist.Normal(mu_selected, sigma_selected).sample(key)

    def predict_posterior_samples(self, X):

        samples = self.get_params_samples()

        beta = samples["beta"]    # (S, K, p)
        pi = samples["pi"]        # (S, K)

        # ---- construir mu ----
        beta_t = jnp.transpose(beta, (0, 2, 1))   # (S, p, K)
        mu = jnp.matmul(X, beta_t)               # (S, n, K)
        mu = jnp.transpose(mu, (1, 0, 2))        # (n, S, K)

        # ---- ponderar por mezcla ----
        pi_expanded = jnp.expand_dims(pi, axis=0)   # (1, S, K)

        y_hat = jnp.sum(mu * pi_expanded, axis=2)   # (n, S)

        return y_hat

    def predict_plugin(self, X):

        params = self.get_plugin_params()

        beta = params["beta"]   # (K, p)
        pi = params["pi"]       # (K,)

        # medias por componente
        mu = X @ beta.T         # (n, K)

        # promedio ponderado
        y_hat = jnp.sum(mu * pi[None, :], axis=1)   # (n,)

        return y_hat# (n,)
```

```python
class BayesianHorseshoeLinearRegression(BayesianLinearRegression):

    def __init__(self,
                 tau_prior=None,
                 lambda_prior=None,
                 **kwargs):

        super().__init__(**kwargs)

        # prior global (tau)
        self.tau_prior = tau_prior if tau_prior is not None else dist.HalfCauchy(1.0)

        # prior local (lambda_j)
        self.lambda_prior = lambda_prior if lambda_prior is not None else dist.HalfCauchy(1.0)

        # no usamos beta_prior base
        self.beta_prior = None


    # =========================================================
    # MODEL
    # =========================================================

    def model(self, X, y=None):

        n, p = X.shape

        # ruido
        sigma = numpyro.sample("sigma", self.sigma_prior)

        # shrinkage global
        tau = numpyro.sample("tau", self.tau_prior)

        # shrinkage local (vector)
        lambda_local = numpyro.sample(
            "lambda",
            self.lambda_prior.expand([p]).to_event(1)
        )

        # varianza por coeficiente
        scale = tau * lambda_local

        # prior sobre beta
        beta = numpyro.sample(
            "beta",
            dist.Normal(0.0, scale).to_event(1)
        )

        mu = jnp.dot(X, beta)

        with numpyro.plate("data", n):
            numpyro.sample("y", dist.Normal(mu, sigma), obs=y)
```

```python
class BayesianContinuousSpikeSlabLinearRegression(BayesianLinearRegression):

    def __init__(self,
                 spike_scale=1e-3,
                 slab_scale_prior=None,
                 z_prior=None,
                 **kwargs):

        super().__init__(**kwargs)

        self.spike_scale = spike_scale
        self.slab_scale_prior = slab_scale_prior if slab_scale_prior is not None else dist.HalfNormal(1.0)
        self.z_prior = z_prior if z_prior is not None else dist.Beta(1.0, 1.0)

        self.beta_prior = None


    # =========================================================
    # MODEL
    # =========================================================

    def model(self, X, y=None):

        n, p = X.shape

        # ruido
        sigma = numpyro.sample("sigma", self.sigma_prior)

        # escala del slab
        slab_scale = numpyro.sample("slab_scale", self.slab_scale_prior)

        # variables continuas (relajan selección)
        z = numpyro.sample(
            "z",
            self.z_prior.expand([p]).to_event(1)
        )

        # varianza interpolada
        scale = jnp.sqrt(
            (1 - z) * (self.spike_scale ** 2) +
            z * (slab_scale ** 2)
        )

        # coeficientes
        beta = numpyro.sample(
            "beta",
            dist.Normal(0.0, scale).to_event(1)
        )

        mu = jnp.dot(X, beta)

        with numpyro.plate("data", n):
            numpyro.sample("y", dist.Normal(mu, sigma), obs=y)
```

```python
from numpyro.infer import Predictive, SVI, Trace_ELBO
from numpyro.infer.autoguide import AutoNormal
from numpyro.optim import Adam
```

```python
hidden_dim = 10
K = 10
batch_size = 200
lr = 9e-3
def bnn_model(X, y=None):

    n, p = X.shape

    W1 = numpyro.sample(
            "W1",
            dist.Normal(0.0, 1.0).expand([p, hidden_dim]).to_event(2)
        )

    b1 = numpyro.sample(
            "b1",
            dist.Normal(0.0, 1.0).expand([hidden_dim]).to_event(1)
        )

    W2 = numpyro.sample(
            "W2",
            dist.Normal(0.0, 1.0).expand([hidden_dim, K]).to_event(2)
        )

    b2 = numpyro.sample(
            "b2",
            dist.Normal(0.0, 1.0).expand([K]).to_event(1)
    )


    with numpyro.plate("data", n, subsample_size=batch_size) as idx:

            X_batch = X[idx]
            y_batch = y[idx] if y is not None else None

            a = jnp.tanh(jnp.dot(X_batch, W1) + b1)

            z = jnp.tanh(jnp.dot(a, W2) + b2)

            numpyro.sample("y", dist.Categorical(logits=z), obs=y_batch)



guide = AutoNormal(bnn_model)

svi = SVI(bnn_model, guide, optim=Adam(lr), loss=Trace_ELBO())

svi_result = svi.run(random.PRNGKey(18081997), 40000, X_train, y_train, progress_bar=True)
```

```python
predictive = Predictive(guide, params=svi_result.params, num_samples=1000)
samples = predictive(random.PRNGKey(1), X_train)
```

```python
W1_samples = samples["W1"]
W2_samples = samples["W2"]
b1_samples = samples["b1"]
b2_samples = samples["b2"]
```
