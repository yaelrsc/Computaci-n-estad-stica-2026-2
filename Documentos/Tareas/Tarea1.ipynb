```python
import torch
import math
```

```python
T1 = torch.distributions.Exponential(1.0)
T2 = torch.distributions.Gamma(2,1/3)
T3 = torch.distributions.Exponential(1.25)
T4 = torch.distributions.LogNormal(0.2, math.sqrt(0.5))
T5 = torch.distributions.Exponential(0.75)
```

```python
n_sim = 100000
S_1 = T1.sample((n_sim,)) + T2.sample((n_sim,))
S_2 = T4.sample((n_sim,)) + T5.sample((n_sim,))
S_3 = T1.sample((n_sim,)) + T3.sample((n_sim,)) + T5.sample((n_sim,))
#T es el max{S_1,S_2,S_3}
```

```python
T = torch.maximum(S_1,torch.maximum(S_2,S_3))
```

```python
T.quantile(torch.tensor(0.95))
```
